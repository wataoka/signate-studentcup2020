## the number of words in a sentence

```python
count_num_words = lambda s: len(s.split(' '))

data = pd.DataFrame([])
data['num_words'] = train_df['description'].apply(count_num_words)

plt.figure(figsize=(20, 10))
sns.countplot('num_words', data=data, color='gray')
```
